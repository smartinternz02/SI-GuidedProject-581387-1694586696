# DATA VISUALIZATION

# import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# get names of all data sets

```python
a=sns.get_dataset_names()
a
```

# loading the dataset

```python
d=sns.load_dataset("car_crashes")
d
```

# Info about the dataset

```python
d.info()
```

# Univariate Analysis

## Box plot to check for outliers 

```python
a=plt.figure()
a.set_figwidth(20)
a.set_figheight(20)
plt.subplot(3,3,1)
sns.boxplot(x="total",data=d)
plt.subplot(3,3,2)
sns.boxplot(x="speeding",data=d)
plt.subplot(3,3,3)
sns.boxplot(x="not_distracted",data=d)
plt.subplot(3,3,4)
sns.boxplot(x="ins_losses",data=d)
plt.subplot(3,3,5)
sns.boxplot(x="ins_premium",data=d)
plt.subplot(3,3,6)
sns.boxplot(x="no_previous",data=d)
```

# Histogram 

```python
sns.histplot(d['ins_premium'], color='blue')
plt.xlabel('Insurance Premium')
plt.ylabel('Frequency')
plt.title('Histogram of Insurance Premium',fontsize=20)
plt.show()
```

```python
#inference is for premium of 800 it has high frequency
```

```python
sns.histplot(d['total'],color='black',kde=True)
plt.title('Histogram of Total',fontsize=30)
```

```python
#when the total is 12.5 the count is maximum i.e 12
```

```python
sns.histplot(d['abbrev'],color='green',kde=True)
plt.title('Histogram of Abbrev',fontsize=30)
```

```python
# for categorical data the histogram displays like this
```

```python
sns.histplot(d['alcohol'],color='red',kde=True)
plt.title('Histogram of Alcohol',fontsize=30)
```

```python
#Maximum of the bar is at count 12
```

# BIVARIATE ANALYSIS

# Heat Map

# Finding correlation for each attribute

```python
a=d.corr()
a
```

```python
sns.heatmap(a,annot=True,cmap="YlGnBu")
```

```python
#Inference: In above heatmap blue indicates extreme values which are positively correlated and green represents 
#negatively correlated.We can get carcrashes more precisely like higher the speeding there is a chance of more 
#likely to have accident.It also tells alcohol intake and carcrashes are directly proportional .
```

# Line plot

```python
a=sns.lineplot(x="speeding",y="alcohol",data=d,color="red")
plt.title("alcohol vs speeding",fontsize=20)
```

```python
# inference is that when alcohol increases speed also increases
#directly proportional
```

```python
sns.lineplot(x="alcohol",y="not_distracted",data=d,color="black")
plt.title("not_distracted vs alcohol",fontsize=20)
```

```python
#inference is when there is less levels of distraction the levels of not getting distracted is higher
```

```python
sns.lineplot(x="ins_premium",y="ins_losses",data=d,color="green")
plt.title("ins_losses vs ins_premium",fontsize=20)
```

```python
#inference is as premium increases ideally losses decreases
```

# Count plot

```python
sns.set_style("whitegrid")  
sns.countplot(x="total", data=d)
plt.title("Count of Total",fontsize=20)
plt.show()
```

```python
# inference is this gives the total count of car crashes
```

```python
sns.set_style("whitegrid")  
sns.countplot(x="not_distracted", data=d)
plt.title("Count of Not Distracted",fontsize=20)
plt.show()
```

```python
#inference is the count of not getting distracted is constant
```

```python
sns.set_style("whitegrid")  
sns.countplot(x="alcohol", data=d)
plt.title("Count of Alcohol",fontsize=20)
plt.show()
```

```python
#inference is the count of alcohol is constant
```
